```python
from bs4 import BeautifulSoup
import requests
```

```python
import re
```

```python
url="https://www.amazon.in/s?k=laptop&rh=n%3A1375424031&ref=nb_sb_noss"
print(url)
```

```python
request=requests.get(url,'html.parse')
print(request)
```

```python
data=BeautifulSoup(request.content)
```

```python
data
```

```python
title=data.find('h2',{'class':'a-size-mini a-spacing-none a-color-base s-line-clamp-2'}).text
title
```

```python
## Lets remove the comma in title. 
title = title.strip().replace(',', '')
title
```

```python
# Replace " With INCH
title = title.strip().replace(' " ', 'Inch')
title
```

```python
title_list=[]
title=data.findAll('h2',{'class':'a-size-mini a-spacing-none a-color-base s-line-clamp-2'})
for i in title:
    title_list.append(i.text)
title_list
```

```python
# Lets remove the comma and "--> inch
title_list_f=[]
for i in title_list:
    i=i.strip().replace(',',' ').replace('" ', 'Inch')
    title_list_f.append(i)
print(title_list_f)
len(title_list)
```

```python
##RATING DATA
rating_list=[]
rating=data.findAll('span',{'class':'a-icon-alt'})
for i in rating:
    rating_list.append(i.text)
rating_list
```

```python
## PRICE
price_list=[]
price=data.findAll('span',{'class':'a-price-whole'})
for i in price:
    price_list.append(i.text)
price_list
```
